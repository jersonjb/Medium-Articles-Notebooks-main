```python
# Importing necessary libraries from transformers and torch
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed
import torch
```

```python
# If using a GPU, ensure torch is set to use it
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
# Set the model ID for BLOOM
model_ID = "bigscience/bloom-1b7"

# Load the pre-trained BLOOM model and tokenizer from Hugging Face
model = AutoModelForCausalLM.from_pretrained(model_ID, use_cache=True)
tokenizer = AutoTokenizer.from_pretrained(model_ID)
```

```python
# Set a seed for reproducibility of results
set_seed(2024)
```

```python
# Define a creative story title to generate text about
story_title = 'An Unexpected Journey Through Time'
prompt = f'This is a creative story about {story_title}.\n'
```

```python
print(prompt)
```

```python
# Tokenize the prompt and map to the appropriate model device
input_ids = tokenizer(prompt, return_tensors="pt").to(0)
```

```python
# Generate text with specified parameters to control creativity and prevent repetition
sample = model.generate(**input_ids, max_length=300,
                        top_k=1, temperature=0.9,
                        repetition_penalty = 2.0)
```

```python
generated_story = tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"])
```

```python
import textwrap
wrapper = textwrap.TextWrapper(width=80)
```

```python
formated_story = wrapper.fill(text=generated_story)
print(formated_story)
```
