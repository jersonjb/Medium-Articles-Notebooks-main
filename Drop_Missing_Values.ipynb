# Drop missing observations from your data

```python
import pandas as pd
import numpy as np
```

```python
candidates_info = {
    'Full_Name':["Aida Kone",np.nan,"Ismael Cisse","Aicha Konate",
                 "Fanta Koumare", "Khalil Cisse"],
    'degree':['Master',np.nan,np.nan, "PhD", "Masters", np.nan],
    'From':["Abidjan",np.nan,np.nan, "Abidjan","Dakar", np.nan],
    'Age':[23,np.nan,19, 30,25, np.nan],
          }

candidates_df = pd.DataFrame(candidates_info) 
candidates_df
```

```python
candidates_df.dropna()
```

```python
candidates_df.dropna(how="all")
```

```python
candidates_df.dropna(thresh=2)
```
